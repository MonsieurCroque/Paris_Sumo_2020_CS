**Calcul du nombre de capteurs**

***Aide à la décision***

On peut arriver à un principe simple. On choisit le véhicule qui donne la distance de Voronoi minimum pour les arrondissements qui nous interesse. On en déduit le nombre de véhicule à équiper en extrapolant à partir du nombre de véhicules dans la simulations et l'importance de la commune en flux sortant.

***Avec Voronoi seulement et un mix de véhicules***

```python
min_vor = data_vor.idxmin(axis = 1)
communes = data_vor.axes[0].tolist()

nb_capteurs = {'car' : 0, 'bike' : 0, 'bus' : 0, 'trucks' : 0}
total_insee = sum(insee_data["NBFLUX_C16_ACTOCC15P"])

for row in range(len(min_vor)):
    nb_capteurs[min_vor[row]] += nb_veh[min_vor[row]] * sum(insee_data[insee_data["CODGEO"] == communes[row]]["NBFLUX_C16_ACTOCC15P"]) / total_insee

nb_capteurs["bus"] /= 20
print(nb_capteurs)
```
